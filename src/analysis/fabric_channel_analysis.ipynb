```python
# Import library yang diperlukan
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

# Fungsi untuk menampilkan channel RGB
def show_rgb_channels(image_path):
    # Baca gambar
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Tidak dapat membaca gambar dari path: {image_path}")
        return
    
    # Konversi dari BGR ke RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Pisahkan channel R, G, B
    r, g, b = cv2.split(image_rgb)
    
    # Buat figure untuk menampilkan gambar
    plt.figure(figsize=(20, 10))
    
    # Gambar asli
    plt.subplot(2, 4, 1)
    plt.imshow(image_rgb)
    plt.title('Original Image Katun (Alami)')
    plt.axis('off')
    
    # Channel Red
    plt.subplot(2, 4, 2)
    plt.imshow(r, cmap='Reds')
    plt.title('Red Channel')
    plt.axis('off')
    
    # Channel Green
    plt.subplot(2, 4, 3)
    plt.imshow(g, cmap='Greens')
    plt.title('Green Channel')
    plt.axis('off')
    
    # Channel Blue
    plt.subplot(2, 4, 4)
    plt.imshow(b, cmap='Blues')
    plt.title('Blue Channel')
    plt.axis('off')
    
    # Histogram untuk channel Red
    plt.subplot(2, 4, 6)
    plt.hist(r.ravel(), bins=256, color='red', alpha=0.7)
    plt.title('Red Histogram')
    plt.xlabel('Intensity Value')
    plt.ylabel('Frequency')
    
    # Histogram untuk channel Green
    plt.subplot(2, 4, 7)
    plt.hist(g.ravel(), bins=256, color='green', alpha=0.7)
    plt.title('Green Histogram')
    plt.xlabel('Intensity Value')
    plt.ylabel('Frequency')
    
    # Histogram untuk channel Blue
    plt.subplot(2, 4, 8)
    plt.hist(b.ravel(), bins=256, color='blue', alpha=0.7)
    plt.title('Blue Histogram')
    plt.xlabel('Intensity Value')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
    
    # Hitung dan tampilkan nilai rata-rata untuk setiap channel
    r_avg = np.mean(r)
    g_avg = np.mean(g)
    b_avg = np.mean(b)
    
    print(f"Red Channel Average: {r_avg:.2f}")
    print(f"Green Channel Average: {g_avg:.2f}")
    print(f"Blue Channel Average: {b_avg:.2f}")
    
    return r_avg, g_avg, b_avg

# Fungsi untuk menampilkan channel NIR
def show_nir_channel(image_path):
    # Baca gambar
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Tidak dapat membaca gambar dari path: {image_path}")
        return
    
    # Jika gambar berwarna, ambil channel pertama (asumsi NIR disimpan di channel pertama)
    if len(image.shape) == 3:
        nir = image[:, :, 0]
        # Konversi dari BGR ke RGB untuk menampilkan gambar asli
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    else:
        nir = image
        image_rgb = image
    
    # Buat figure untuk menampilkan gambar
    plt.figure(figsize=(15, 5))
    
    # Gambar asli
    plt.subplot(1, 3, 1)
    plt.imshow(image_rgb)
    plt.title('Original Image Katun (Alami)')
    plt.axis('off')
    
    # Channel NIR
    plt.subplot(1, 3, 2)
    plt.imshow(nir, cmap='viridis')
    plt.title('NIR Channel')
    plt.axis('off')
    plt.colorbar()
    
    # Histogram untuk channel NIR
    plt.subplot(1, 3, 3)
    plt.hist(nir.ravel(), bins=256, color='purple', alpha=0.7)
    plt.title('NIR Histogram')
    plt.xlabel('Intensity Value')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
    
    # Hitung dan tampilkan nilai rata-rata untuk channel NIR
    nir_avg = np.mean(nir)
    print(f"NIR Channel Average: {nir_avg:.2f}")
    
    return nir_avg

# Contoh penggunaan
if __name__ == "__main__":
    # Ganti dengan path gambar Anda
    rgb_image_path = r"D:\Semester7\CV\identifikasi_kain\data\raw\RGB\katun\katun_rgb_01.jpg"
    nir_image_path = r"D:\Semester7\CV\identifikasi_kain\data\raw\NIR\katun\katun_nir_01.jpg"
    
    # Pastikan file exists
    if os.path.exists(rgb_image_path):
        print("Analysing RGB image:")
        show_rgb_channels(rgb_image_path)
    else:
        print(f"RGB image not found at: {rgb_image_path}")
    
    if os.path.exists(nir_image_path):
        print("\nAnalysing NIR image:")
        show_nir_channel(nir_image_path)
    else:
        print(f"NIR image not found at: {nir_image_path}")
```

# Validasi Ekstraksi Fitur dari Gambar RGB dan NIR

Anda benar untuk mempertanyakan validitas ekstraksi fitur. Mari saya jelaskan metode ekstraksi fitur yang digunakan dalam kode dan bagaimana memvalidasi hasilnya.

## Metode Ekstraksi Fitur yang Digunakan

Kode yang saya berikan menggunakan dua pendekatan ekstraksi fitur:

### 1. Ekstraksi Fitur Warna Dasar (RGB)
- **Channel Separation**: Memisahkan gambar menjadi tiga channel warna (Red, Green, Blue)
- **Intensity Averaging**: Menghitung nilai rata-rata intensitas untuk setiap channel
- **Histogram Analysis**: Menganalisis distribusi nilai intensitas melalui histogram

### 2. Ekstraksi Fitur NIR (Near-Infrared)
- **Channel Extraction**: Mengambil channel Near-Infrared (biasanya channel pertama pada gambar multispektral) Red sebagai proxy NIR
- **Intensity Analysis**: Menganalisis nilai intensitas NIR


## Cara Memvalidasi Ekstraksi Fitur

### 1. Validasi Visual
- **Channel Separation**: Pastikan setiap channel (R, G, B, NIR) ditampilkan dengan benar
- **Histogram Analysis**: Periksa distribusi nilai intensitas pada histogram

### 2. Validasi Statistik
- **Nilai Rata-rata**: Bandingkan nilai rata-rata dengan nilai yang diharapkan
- **Standar Deviasi**: Periksa variasi nilai intensitas
- **Rentang Nilai**: Pastikan nilai berada dalam rentang yang wajar (0-255 untuk gambar 8-bit)

### 3. Validasi Konsistensi
- Bandingkan hasil ekstraksi fitur dari beberapa gambar dengan karakteristik serupa
- Pastikan perbedaan antara katun dan polyester konsisten

### 4. Validasi dengan Tool Eksternal
- Bandingkan hasil dengan software pengolah gambar seperti Photoshop atau GIMP
- Gunakan library alternatif seperti scikit-image untuk memverifikasi hasil

## Kelebihan Metode Ini

1. **Transparan**: Semua langkah ekstraksi fitur dapat dilihat dan diverifikasi
2. **Reproducible**: Hasil yang sama akan diperoleh dengan gambar dan parameter yang sama
3. **Interpretable**: Fitur yang diekstraksi mudah dipahami (nilai intensitas rata-rata)
4. **Validatable**: Dapat divalidasi dengan berbagai metode

## Catatan Penting

1. Pastikan format gambar sesuai dengan ekspektasi:
   - Gambar RGB harus memiliki 3 channel
   - Gambar NIR bisa berupa grayscale atau multispektral

2. Jika gambar NIR Anda memiliki format khusus, Anda mungkin perlu menyesuaikan kode untuk mengekstrak channel NIR yang benar.

3. Untuk analisis yang lebih mendalam, pertimbangkan untuk mengekstrak fitur tambahan seperti:
   - Texture features (GLCM, LBP)
   - Statistical features (skewness, kurtosis)
   - Frequency domain features (FFT, Wavelet)

Dengan kode dan validasi ini, Anda dapat yakin bahwa ekstraksi fitur dilakukan dengan benar dan hasilnya dapat diandalkan untuk analisis lebih lanjut.
