```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import VGG16
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import cv2
import joblib

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define paths
BASE_PATH = r'D:\Semester7\CV\identifikasi_kain\data\raw'
RGB_PATH = os.path.join(BASE_PATH, 'RGB')
NIR_PATH = os.path.join(BASE_PATH, 'NIR')

# Define categories
NATURAL_FABRICS = ['katun', 'sutra']
SYNTHETIC_FABRICS = ['polyester', 'rayon']
ALL_FABRICS = NATURAL_FABRICS + SYNTHETIC_FABRICS

# Define image size
IMG_SIZE = (224, 224)

# Function to load and preprocess images
def load_and_preprocess_image(image_path, is_nir=False):
    """Load and preprocess image for the model"""
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Failed to load image: {image_path}")
        
    if is_nir:
        # For NIR images, ensure they're single channel and replicate to 3 channels
        if len(img.shape) == 3:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    else:
        # For RGB images, convert from BGR to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Resize image
    img = cv2.resize(img, IMG_SIZE)
    # Normalize pixel values
    img = img.astype('float32') / 255.0
    return img

# Function to load dataset
def load_dataset():
    """Load and prepare the dataset"""
    rgb_images = []
    nir_images = []
    fabric_types = []
    categories = []
    
    # Loop through each fabric type
    for fabric in ALL_FABRICS:
        rgb_folder = os.path.join(RGB_PATH, fabric)
        nir_folder = os.path.join(NIR_PATH, fabric)
        
        # Check if folders exist
        if not os.path.exists(rgb_folder):
            print(f"Warning: RGB folder not found: {rgb_folder}")
            continue
        if not os.path.exists(nir_folder):
            print(f"Warning: NIR folder not found: {nir_folder}")
            continue
            
        # Get list of images
        rgb_files = [f for f in os.listdir(rgb_folder) if f.lower().endswith(('.jpg', '.png', '.jpeg', '.JPG', '.PNG', '.JPEG'))]
        nir_files = [f for f in os.listdir(nir_folder) if f.lower().endswith(('.jpg', '.png', '.jpeg', '.JPG', '.PNG', '.JPEG'))]

        print(f"Processing {fabric} - Found {len(rgb_files)} RGB images")
        print(f"Processing {fabric} - Found {len(nir_files)} NIR images")
        
        # Process only RGB files and find NIR pairs
        for rgb_file in rgb_files:
            try:
                rgb_path = os.path.join(rgb_folder, rgb_file)
                rgb_img = load_and_preprocess_image(rgb_path, is_nir=False)

                # Find NIR pair (replace pattern _rgb_ -> _nir_)
                nir_file = rgb_file.replace("_rgb_", "_nir_")
                nir_path = os.path.join(nir_folder, nir_file)

                if not os.path.exists(nir_path):
                    print(f"Warning: NIR pair not found for {rgb_file}")
                    continue

                nir_img = load_and_preprocess_image(nir_path, is_nir=True)

                rgb_images.append(rgb_img)
                nir_images.append(nir_img)
                fabric_types.append(fabric)
                categories.append('alami' if fabric in NATURAL_FABRICS else 'sintetis')

            except Exception as e:
                print(f"Error processing RGB {rgb_file}: {str(e)}")
                continue
    
    # Check if any images were loaded
    if len(rgb_images) == 0:
        raise ValueError("No images were loaded. Please check the dataset paths and image files.")
    
    print(f"\nDataset Summary:")
    print(f"Total pairs loaded: {len(rgb_images)}")
    for fabric in ALL_FABRICS:
        count = fabric_types.count(fabric)
        print(f"{fabric}: {count} pairs")
    
    return np.array(rgb_images), np.array(nir_images), np.array(fabric_types), np.array(categories)

# Load the dataset
print("Loading and preprocessing images...")
rgb_images, nir_images, fabric_types, categories = load_dataset()

# Convert labels to numerical format
le_fabric = LabelEncoder()
le_category = LabelEncoder()

y_fabric = le_fabric.fit_transform(fabric_types)
y_category = le_category.fit_transform(categories)

# Convert to categorical
y_fabric = tf.keras.utils.to_categorical(y_fabric)
y_category = tf.keras.utils.to_categorical(y_category)

# Split the dataset
X_rgb_train, X_rgb_test, X_nir_train, X_nir_test, \
y_fabric_train, y_fabric_test, y_category_train, y_category_test = train_test_split(
    rgb_images, nir_images, y_fabric, y_category, 
    test_size=0.2, random_state=42, stratify=y_fabric
)

print("\nDataset shapes:")
print(f"RGB Training: {X_rgb_train.shape}")
print(f"NIR Training: {X_nir_train.shape}")
print(f"Fabric Types: {y_fabric_train.shape}")
print(f"Categories: {y_category_train.shape}")

# Print label encoding mapping
print("\nLabel Encoding:")
print("Fabric types:", dict(zip(le_fabric.classes_, le_fabric.transform(le_fabric.classes_))))
print("Categories:", dict(zip(le_category.classes_, le_category.transform(le_category.classes_))))

# Build a simpler model to prevent overfitting
def build_simple_model(input_shape):
    """Build a simpler model with more regularization to prevent overfitting"""
    # Input layers
    rgb_input = layers.Input(shape=input_shape, name='rgb_input')
    nir_input = layers.Input(shape=input_shape, name='nir_input')
    
    # Shared convolutional base
    def create_conv_base():
        model = tf.keras.Sequential([
            layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
            layers.MaxPooling2D((2, 2)),
            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.MaxPooling2D((2, 2)),
            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.GlobalAveragePooling2D()
        ])
        return model
    
    conv_base = create_conv_base()
    
    # Process both inputs
    rgb_features = conv_base(rgb_input)
    nir_features = conv_base(nir_input)
    
    # Combine features
    combined = layers.Concatenate()([rgb_features, nir_features])
    
    # Add dropout and regularization
    x = layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(combined)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(32, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    
    # Output layers
    fabric_output = layers.Dense(len(ALL_FABRICS), activation='softmax', name='fabric_output')(x)
    category_output = layers.Dense(2, activation='softmax', name='category_output')(x)
    
    # Create model
    model = models.Model(
        inputs=[rgb_input, nir_input],
        outputs=[fabric_output, category_output]
    )
    
    return model

# Create the simple model
input_shape = (*IMG_SIZE, 3)
model = build_simple_model(input_shape)

# Calculate class weights for fabric types to handle any imbalance
fabric_classes = np.argmax(y_fabric_train, axis=1)
fabric_class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(fabric_classes),
    y=fabric_classes
)
fabric_class_weight_dict = dict(enumerate(fabric_class_weights))
print("Class weights:", fabric_class_weight_dict)

# Compile model with class weights
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(
    optimizer=optimizer,
    loss={
        'fabric_output': 'categorical_crossentropy',
        'category_output': 'categorical_crossentropy'
    },
    loss_weights={
        'fabric_output': 1.0,
        'category_output': 0.3
    },
    metrics={
        'fabric_output': 'accuracy',
        'category_output': 'accuracy'
    }
)

# Enhanced callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=8,
    min_lr=1e-7,
    verbose=1
)

model_checkpoint = ModelCheckpoint(
    'best_model.h5',
    monitor='val_fabric_output_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

# Apply data augmentation to the training data manually
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create augmented datasets
def augment_data(images, labels, num_augmentations=3):
    datagen = ImageDataGenerator(
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        zoom_range=0.1,
        brightness_range=[0.9, 1.1],
        fill_mode='nearest'
    )
    
    augmented_images = []
    augmented_labels = []
    
    for i in range(len(images)):
        # Original image
        augmented_images.append(images[i])
        augmented_labels.append(labels[i])
        
        # Augmented images
        img = images[i]
        img = img.reshape((1,) + img.shape)
        
        # Create a generator for the current image
        aug_gen = datagen.flow(img, batch_size=1)
        
        for j in range(num_augmentations):
            # Use next() function
            augmented_img = next(aug_gen)[0]
            augmented_images.append(augmented_img)
            augmented_labels.append(labels[i])
    
    return np.array(augmented_images), np.array(augmented_labels)

# Augment RGB and NIR data separately
print("Augmenting RGB training data...")
X_rgb_augmented, y_fabric_augmented = augment_data(X_rgb_train, y_fabric_train)

print("Augmenting NIR training data...")
X_nir_augmented, y_category_augmented = augment_data(X_nir_train, y_category_train)

# Check if we have enough memory for the augmented data
print(f"Original RGB data shape: {X_rgb_train.shape}")
print(f"Augmented RGB data shape: {X_rgb_augmented.shape}")
print(f"Original NIR data shape: {X_nir_train.shape}")
print(f"Augmented NIR data shape: {X_nir_augmented.shape}")

# Prepare validation data
val_data = ([X_rgb_test, X_nir_test], [y_fabric_test, y_category_test])

# Training parameters
EPOCHS = 50
BATCH_SIZE = 8

# Train with augmented data
print("Training improved model with augmented data...")
history = model.fit(
    [X_rgb_augmented, X_nir_augmented],
    [y_fabric_augmented, y_category_augmented],
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=val_data,
    callbacks=[early_stopping, reduce_lr, model_checkpoint],
    verbose=1
)

# Load the best model
model.load_weights('best_model.h5')

# Plot training history
plt.figure(figsize=(15, 10))

# Plot accuracy
plt.subplot(2, 2, 1)
plt.plot(history.history['fabric_output_accuracy'], label='Fabric Training Accuracy')
plt.plot(history.history['val_fabric_output_accuracy'], label='Fabric Validation Accuracy')
plt.title('Fabric Classification Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(history.history['category_output_accuracy'], label='Category Training Accuracy')
plt.plot(history.history['val_category_output_accuracy'], label='Category Validation Accuracy')
plt.title('Category Classification Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(2, 2, 3)
plt.plot(history.history['fabric_output_loss'], label='Fabric Training Loss')
plt.plot(history.history['val_fabric_output_loss'], label='Fabric Validation Loss')
plt.title('Fabric Classification Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(history.history['category_output_loss'], label='Category Training Loss')
plt.plot(history.history['val_category_output_loss'], label='Category Validation Loss')
plt.title('Category Classification Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluate the model
print("Evaluating model on test data...")
test_results = model.evaluate(
    [X_rgb_test, X_nir_test],
    [y_fabric_test, y_category_test],
    verbose=1
)

print(f"Test Loss: {test_results[0]}")
print(f"Fabric Type Accuracy: {test_results[3]}")
print(f"Category Accuracy: {test_results[4]}")

# Make predictions
print("Making predictions on test data...")
y_fabric_pred, y_category_pred = model.predict([X_rgb_test, X_nir_test])

# Convert predictions to class labels
y_fabric_pred_classes = np.argmax(y_fabric_pred, axis=1)
y_fabric_true_classes = np.argmax(y_fabric_test, axis=1)
y_category_pred_classes = np.argmax(y_category_pred, axis=1)
y_category_true_classes = np.argmax(y_category_test, axis=1)

# Convert numerical labels back to original text labels
fabric_pred_text = le_fabric.inverse_transform(y_fabric_pred_classes)
fabric_true_text = le_fabric.inverse_transform(y_fabric_true_classes)

category_pred_text = le_category.inverse_transform(y_category_pred_classes)
category_true_text = le_category.inverse_transform(y_category_true_classes)

# Print classification reports
print("Fabric Type Classification Report:")
print(classification_report(
    y_fabric_true_classes,
    y_fabric_pred_classes,
    target_names=le_fabric.classes_
))

print("\nCategory Classification Report:")
print(classification_report(
    y_category_true_classes,
    y_category_pred_classes,
    target_names=le_category.classes_
))

# Create confusion matrices
fabric_cm = confusion_matrix(fabric_true_text, fabric_pred_text, labels=ALL_FABRICS)
category_cm = confusion_matrix(category_true_text, category_pred_text, labels=['alami', 'sintetis'])

# Plot confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Fabric type confusion matrix
sns.heatmap(fabric_cm, annot=True, fmt='d', cmap='Blues', xticklabels=ALL_FABRICS, yticklabels=ALL_FABRICS, ax=axes[0])
axes[0].set_title('Fabric Type Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('True')

# Category confusion matrix
sns.heatmap(category_cm, annot=True, fmt='d', cmap='Blues', xticklabels=['alami', 'sintetis'], yticklabels=['alami', 'sintetis'], ax=axes[1])
axes[1].set_title('Category Confusion Matrix')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('True')

plt.tight_layout()
plt.show()

# Function to predict fabric type and category for new images
def predict_fabric_type(rgb_image_path, nir_image_path):
    """Predict fabric type and category for a pair of RGB and NIR images"""
    # Load and preprocess images
    rgb_img = load_and_preprocess_image(rgb_image_path, is_nir=False)
    nir_img = load_and_preprocess_image(nir_image_path, is_nir=True)
    
    # Expand dimensions for batch
    rgb_img = np.expand_dims(rgb_img, axis=0)
    nir_img = np.expand_dims(nir_img, axis=0)
    
    # Make prediction
    fabric_pred, category_pred = model.predict([rgb_img, nir_img])
    
    # Get predicted classes
    fabric_type = le_fabric.classes_[np.argmax(fabric_pred[0])]
    category = le_category.classes_[np.argmax(category_pred[0])]
    
    # Get confidence scores
    fabric_confidence = np.max(fabric_pred[0]) * 100
    category_confidence = np.max(category_pred[0]) * 100
    
    return {
        'fabric_type': fabric_type,
        'category': category,
        'fabric_confidence': fabric_confidence,
        'category_confidence': category_confidence
    }

# Save the model and necessary components
save_dir = r'D:\Semester7\CV\identifikasi_kain\outputs\models\model_input_gambar_v2'
os.makedirs(save_dir, exist_ok=True)

# Save the model
model.save(os.path.join(save_dir, 'fabric_classifier_model.h5'))

# Save the label encoders
joblib.dump(le_fabric, os.path.join(save_dir, 'label_encoder_fabric.joblib'))
joblib.dump(le_category, os.path.join(save_dir, 'label_encoder_category.joblib'))

print(f"\nModel and components saved in: {save_dir}")

# Example usage of prediction function
print("\nExample prediction:")
try:
    # Find example images for testing
    example_rgb_path = os.path.join(RGB_PATH, ALL_FABRICS[0], os.listdir(os.path.join(RGB_PATH, ALL_FABRICS[0]))[0])
    example_nir_path = example_rgb_path.replace("RGB", "NIR").replace("_rgb_", "_nir_")
    
    if os.path.exists(example_rgb_path) and os.path.exists(example_nir_path):
        result = predict_fabric_type(example_rgb_path, example_nir_path)
        print(f"Predicted Fabric: {result['fabric_type']} (Confidence: {result['fabric_confidence']:.2f}%)")
        print(f"Predicted Category: {result['category']} (Confidence: {result['category_confidence']:.2f}%)")
    else:
        print("Example images not found for testing prediction function")
except Exception as e:
    print(f"Error testing prediction function: {e}")
```
